```python
from relaylab import equipment as eqp, signals as sg, record, characteristics, functions as func
from relaylab.signals import color
import plotly.graph_objects as go
import numpy as np
```

### Создание сигнала и расчет действующего значения

```python
i = sg.transient(name='I', val_load=1, fi_load=0, val_fault=10, fi_fault=-90, tau=0.03, tfault=0.04, tmax=0.12,
               f=50, Fs=1200)
i_dft = sg.DFT(i)
i_rms = sg.RMS(i)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(i.plt_data, mode="lines+markers", line=dict(color=color.orange)))
fig.add_trace(go.Scatter(i_dft.abs().plt_data, line=dict(color=color.blue)))
fig.add_trace(go.Scatter(i_rms.plt_data, line=dict(color=color.grey_dark)))

fig.update_layout(template='plotly_white', height=500, 
                  shapes=[dict(type= 'line', line= dict(color=color.grey, dash='dash'), 
                               yref= 'y', y0= 8, y1= 8, 
                               xref= 'paper', x0= 0, x1= 1)]
                 )
```
