```python
from relaylab import equipment as eqp, signals as sg, record, characteristics, functions as func
from relaylab.signals import color
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

#### Создание вручную

```python
t = np.arange(0, 0.04, 1/1200)
val = np.sqrt(2) * np.sin(2*np.pi*50*t)
sin_signal = sg.AnalogSignal(name='sin', val=val, Fs=1200)
```

Сигнал создан, теперь можно получить следующие его атрибуты:

```python
sin_signal.val
```

```python
sin_signal.time
```

```python
sin_signal.name
```

```python
sin_signal.Fs
```

#### Создание с помощью втроенной функции:

```python
sin_signal2 = sg.sin(1, 30, name='sin2', tmax=0.04, Fs=1200)
```

#### Построим график сигнала

```python
signal_sum = sin_signal + sin_signal2

fig = go.Figure()
fig.add_trace(go.Scatter(name=sin_signal.name, x=sin_signal.time, y=sin_signal.val, line=dict(color=color.blue)))
fig.add_trace(go.Scatter(sin_signal2.plt_data, line=dict(color=color.red)))
fig.add_trace(go.Scatter(signal_sum.plt_data, line=dict(color=color.green)))

fig.update_layout(template='plotly_white')
```

#### Расчет комплексного значения

```python
fig = go.Figure()
fig.add_trace(go.Scatter(sin_signal2.plt_data, line=dict(color=color.blue)))
fig.add_trace(go.Scatter(sin_signal2.dft_abs().plt_data, line=dict(color=color.red)))
fig.update_layout(template='plotly_white')
```
